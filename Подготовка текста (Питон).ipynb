Беляев "Человек, потерявший лицо". Обработка текста

```python
with open('1929_Belyaev_Chelovek, poteryavshij lico.txt', 'r') as file: # r означает, что вы передаете файл на чтение
    Belyaev_Chelovek = file.read()
```

```python
Belyaev_Chelovek
```

```python
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
Belyaev_Chelovek_tokenized = word_tokenize(Belyaev_Chelovek.lower())
```

```python
Belyaev_Chelovek_tokenized
```

```python
Belyaev_Chelovek_tokenized_no_punct = [word for word in Belyaev_Chelovek_tokenized if word[0].isalpha()]
```

```python
len(Belyaev_Chelovek_tokenized_no_punct)
```

```python
from nltk import FreqDist
```

```python
word_freqs_Belyaev_Chelovek = FreqDist(Belyaev_Chelovek_tokenized_no_punct)
```

```python
word_freqs_Belyaev_Chelovek
```

```python
 keys = word_freqs_Belyaev_Chelovek.keys() 
 values = word_freqs_Belyaev_Chelovek.values()
```

```python
belyaev = pd.DataFrame({'Title': 'Человек, потерявший лицо', 'Word': keys, 'n': values, 'nwords': 22394})
```

```python
belyaev
```

```python
belyaev.to_csv('belyaev.csv') 
```
